##**We can use FDNY Dataset to investigate at what time of the day people cook most.**

```python
cooking_fire = nyfd_data.loc[nyfd_data['INCIDENT_TYPE_DESC'] == '113 - Cooking fire, confined to container']
len(cooking_fire)
cooking_fire.head()
```

```python
cooking_fire['INCIDENT_DATE'], cooking_fire['INCIDENT_TIME'], cooking_fire['HOUR_CLOCK'] = zip(*cooking_fire['INCIDENT_DATE_TIME'].map(lambda x: x.split(' ')))
```

```python
cooking_fire.head()
```

```python
cooking_fire['INCIDENT_HOUR'], cooking_fire['INCIDENT_MINS'], cooking_fire['INCIDENT_SECONDS'] = zip(*cooking_fire['INCIDENT_TIME'].map(lambda x: x.split(':')))
cooking_fire.head()
```

```python
hour_max_val = zip(cooking_fire['INCIDENT_HOUR'], cooking_fire['HOUR_CLOCK'])
print(statistics.mode(hour_max_val))

# SO THE TIME OF DAY PEOPLE COOK MOST IS 18TH HOUR I.E. 6PM 
```
